# **Implementar y entrenar una regresión logística**

```python
def sigmoid(z):
  return 1/(1+ np.exp(-z))
```

```python
def cost_function(y, y_hat, eps=1e-15):
    m = y.shape[0]
    y_hat = np.clip(y_hat, eps, 1 - eps)
    cost = (-1/m) * np.sum(y * np.log(y_hat) + (1-y) * np.log(1-y_hat))
    return cost
```

```python
def predict(x, w, b):
    return np.round(sigmoid(x.dot(w) + b))
```

```python
def predict_proba(x, w, b):
  z = x.dot(w) + b
  return sigmoid(z)
```

```python
def get_gradients(x, y_train, y_pred, w, b):
    m = y_train.shape[0]
    dz = y_pred - y_train
    dw = (1/m) * x.T.dot(dz)
    db = (1/m) * np.sum(dz)
    return dw, db
```

```python
def update_params(w, b, dw, db, learning_rate):
  w  -= learning_rate * dw
  b  -= learning_rate * db
  return w, b
```

```python
def reg_log(x, y, niter, learning_rate):
  w = np.zeros(8)
  b = 0
  for i in range(niter):

    y_pred = predict_proba(x, w, b)

    dw, db = get_gradients(x, y, y_pred, w, b)

    w, b = update_params(w, b, dw, db, learning_rate)

    print(f"Costo: {cost_function(y, y_pred)}")

  return w, b
```

```python
#Se muestra el costo en cada iteración
w, b = reg_log(X_train, y_train, niter=100, learning_rate=0.1)
```

```python
costo = cost_function(y_train,predict_proba(X_train, w, b))
print("Las pendientes obtenidas por el modelo son: ", w)
print("Con un costo final en el conjunto de entrenamiento de: ", costo)
```

<h2><b>Implementar una regresión lineal con descenso del gradiente con las siguientes funciones</b></h2>

```python
#Para este caso se utiliza el mse como medida de rendimiento
def cost_function(y, y_hat):
  m = len(y)
  cost = (1/(2*m)) * np.sum((y_hat - y)**2)
  return cost
```

```python
def predict(x, w, b):
  #y_hat = np.clip(y_hat, 1e-15, 1 - 1e-15
  x[x == 0] = 1e-15
  return x.dot(w) + b
```

```python
def get_gradients(x, y, y_hat, w, b):
  m = len(y)
  dw = (1/m) * x.T.dot( (y_hat - y))
  db = (1/m) * np.sum(y_hat - y)
  return dw, db
```

```python
def update_params(w, b, dw, db, learning_rate):
  w  -= learning_rate * dw
  b  -= learning_rate * db
  return w, b
```

```python
def reg_lineal(x, y, niter, learning_rate):
  w = np.zeros(8)
  b = 0
  for i in range(niter):
      y_hat = predict(x, w, b)
      dw, db = get_gradients(x, y, y_hat, w, b)
      w, b = update_params(w, b, dw, db, learning_rate)
      print(f"Costo: {cost_function(y, y_hat)}")

  return w, b
```

```python
#Se muestra el mse en cada iteración
w, b =  reg_lineal(X_train, y_train, 100, 0.1)
```

```python
print("Las pendientes para el modelos son: ", w)
print("El punto de corte es: ", b)
```
